```python
# importing libraries
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df=pd.read_csv('C:/Users/asthe/Downloads/OfficialNominalCatches/OfficialNominalCatches_08May2019/ICESCatchDataset2006-2017(1table).csv')
end=pd.read_csv('C:/Users/asthe/Downloads/redlist_species_data/assessments.csv')
xls=pd.ExcelFile('C:/Users/asthe/Downloads/OfficialNominalCatches/OfficialNominalCatches_08May2019/ICESCatchDataset2006-2017.xlsx')
Species_Key=pd.read_excel(xls, 'Species')
Country_Key=pd.read_excel(xls, 'Country')
Tax=pd.read_csv('C:/Users/asthe/Downloads/redlist_species_data/taxonomy.csv')
```

```python
DF=pd.merge(df, Species_Key, how='left', left_on='Species', right_on='FAO_code')
DF=pd.merge(DF, Country_Key, how='left', left_on='Country', right_on='Code')
DF=pd.merge(DF, end, left_on='Latin species name', right_on='scientificName')
df=pd.merge(DF, Tax, how='left', on='scientificName')
```

```python
df=df.melt(id_vars=['Species',
 'Area',
 'Units',
 'Country_x',
 'FAO_code',
 'Latin species name',
 'English',
 'Spanish',
 'French',
 'Code',
 'Country_y',
 'assessmentId',
 'internalTaxonId_x',
 'scientificName',
 'redlistCategory',
 'redlistCriteria',
 'yearPublished',
 'assessmentDate',
 'criteriaVersion',
 'language',
 'rationale',
 'habitat',
 'threats',
 'population',
 'populationTrend',
 'range',
 'useTrade',
 'systems',
 'conservationActions',
 'realm',
 'yearLastSeen',
 'possiblyExtinct',
 'possiblyExtinctInTheWild',
 'scopes',
 'internalTaxonId_y',
 'kingdomName',
 'phylumName',
 'orderName',
 'className',
 'familyName',
 'genusName',
 'speciesName',
 'infraType',
 'infraName',
 'infraAuthority',
 'subpopulationName',
 'authority',
 'taxonomicNotes'], value_vars=['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'], var_name="Year", value_name="Fish Catch (Tonnes Live Weight)")
```
